```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
flights = sns.load_dataset('flights')
```

```python
# Explore the dataset
print(flights.head())
```

```python

# Perform data analysis
# 1. Summary statistics
print(flights.describe())
```

```python
sns.pairplot(flights)
```

```python
sns.heatmap(flights.corr(),annot=True,vmin=-1)
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='year', data=flights)
plt.title('Number of Flights per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Visualize the average number of passengers per month
plt.figure(figsize=(10, 6))
sns.lineplot(x='month', y='passengers', data=flights)
plt.title('Average Number of Passengers per Month')
plt.xlabel('Month')
plt.ylabel('Average Number of Passengers')
plt.show()
```

```python
# 4. Visualize the correlation between the number of passengers and the year
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='passengers', data=flights)
plt.title('Correlation between Number of Passengers and Year')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python

# 5. Visualize the distribution of the number of passengers
plt.figure(figsize=(10, 6))
sns.histplot(flights['passengers'], kde=True)
plt.title('Distribution of Number of Passengers')
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
plt.show()
```
